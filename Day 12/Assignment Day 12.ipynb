## Non - parametric test

### 1. Wilcoxson Sign Test -> To compare the 2 paired samples

```python
from scipy.stats import wilcoxon

d1 = dataset.StockOptionLevel

d2 = dataset.TrainingTimesLastYear

stats,p = wilcoxon( d1 , d2 )

print(stats,p)

print('\nCorrelation :',stats,'\np-value :',p)

if p >= 0.5 :
    print("\nAccept null hypothesis")
elif p < 0.5 :
    print("\nReject null hypothesis \n")
    
print('H0 -> There is no significant difference in StockOptionLevel and TrainingTimesLastYear')
print('HA -> There is significant difference in StockOptionLevel and TrainingTimesLastYear')
```

### 2. Freidman Test -> To compare more than 2 paired samples

```python
from scipy.stats import friedmanchisquare

d1 = dataset.StockOptionLevel

d2 = dataset.TrainingTimesLastYear

d3 = dataset.YearsSinceLastPromotion

stat,p = friedmanchisquare(d1,d2,d3)

print(stats,p)

print('\nCorrelation :',stats,'\np-value :',p)

if p >= 0.5 :
    print("\nAccept null hypothesis")
elif p < 0.5 :
    print("\nReject null hypothesis ")
    
print('\nH0 -> There is no significant difference in StockOptionLevel ,TrainingTimesLastYear and YearsSinceLastPromotion')
print('HA -> There is significant difference in StockOptionLevel , TrainingTimesLastYear and YearsSinceLastPromotion')
```

### 3. Mann whitney Test -> To compare 2 independent samples

```python
from scipy.stats import mannwhitneyu

a1 = dataset.JobLevel

a2 = dataset.Education

stat,p = mannwhitneyu(a1,a2)

print(stat,p)

print('\nCorrelation :',stats,'\np-value :',p)

if p >= 0.5 :
    print("\nAccept null hypothesis")
elif p < 0.5 :
    print("\nReject null hypothesis ")
    
print('\nH0 -> There is no significant difference in JobLevel and Education')
print('HA -> There is significant difference in JobLevel and Education')
```

### 4. Krushkal Wallis Test -> To compare more than 2 independent samples

```python
from scipy.stats import kruskal

a3 = dataset.Attrition

stat,p=kruskal(a1,a2,a3)

print(stat,p)

print('\nCorrelation :',stats,'\np-value :',p)

if p >= 0.5 :
    print("\nAccept null hypothesis")
elif p < 0.5 :
    print("\nReject null hypothesis ")

print('\nH0 -> There is no significant difference in JobLevel, Education and Attrition')
print('HA -> There is significant difference in JobLevel, Education and Attrition')
```

### 5. Chi square test -> To check the dependency of the variables. Variables should be categorical

```python
from scipy.stats import chi2_contingency

b1 = dataset.Attrition
b2 = dataset.Gender

chitable = pd.crosstab( b1,b2)

chitable
```

```python
stats,p,dof,expeted = chi2_contingency(chitable)

print(stats,p)

print('\nCorrelation :',stats,'\np-value :',p)

if p >= 0.5 :
    print("\nAccept null hypothesis")
elif p < 0.5 :
    print("\nReject null hypothesis ")
    
print('\nH0 -> There is no dependency b/w Attrition and Gender')
print('Ha -> There is dependency b/w Attrition and Gender')
```

# Parametric test

### 1. One Sample Test -> To compare the sample mean with population mean


```python
from scipy.stats import ttest_1samp

m1 = dataset.MonthlyIncome

stat,p = ttest_1samp( m1 , 65)

print(stats,p)

print('\nCorrelation :',stats,'\np-value :',p)

if p >= 0.5 :
    print("\nAccept null hypothesis")
elif p < 0.5 :
    print("\nReject null hypothesis ")

print('\nH0 -> There is no significant difference of the mean of Employee MonthlyIncome against population mean = 65')
print('H1 -> There is significant difference of the mean of Employee MonthlyIncome against population mean = 65')
```

### 2.Two sample paired t test -> To compare the mean of 2 paired samples

```python
from scipy.stats import ttest_rel

p1 = dataset.YearsSinceLastPromotion
p2 = dataset.YearsAtCompany

stat,p=ttest_rel(p1,p2)

print(stats,p)

print('\nCorrelation :',stats,'\np-value :',p)

if p >= 0.5 :
    print("\nAccept null hypothesis")
elif p < 0.5 :
    print("\nReject null hypothesis ")

print('\nH0 -> There is no significant difference in the mean of YearsSinceLastPromotion and YearsAtCompany')
print('H1 -> There is significant difference in the mean of YearsSinceLastPromotion and YearsAtCompany')
```

### 3. Two Sample independent T-Test -> To compare the mean of 2 independent samples
